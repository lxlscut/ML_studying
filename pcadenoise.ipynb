```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
digits = datasets.load_digits()
x = digits.data
y = digits.target
x.shape
y.shape
digits
```

```python
x_noise = x + np.random.normal(0,4,size=x.shape)
```

```python
example_digits = x_noise[y==0,:][:10]
example_digits.shape
for num in range(1,10):
    X_num = x_noise[y==num,:][:10]
    example_digits = np.row_stack([example_digits,X_num])
```

```python
def plot_digits(data):
    fig,axes = plt.subplots(10,10,figsize = (10,10),
                            subplot_kw={'xticks':[],'yticks':[]},
                            gridspec_kw=dict(hspace = 0.1,wspace = 0.1))
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),cmap='binary',interpolation='nearest',
                  clim = (0,16))
```

```python
plot_digits(example_digits)
plt.show()

```

```python
from sklearn.decomposition import PCA
pca = PCA(0.4)
pca.fit(example_digits)
example_digits_reduction = pca.transform(example_digits)
example_digits_denoise = pca.inverse_transform(example_digits_reduction)
```

```python
plot_digits(example_digits_denoise)
plt.show()

```
