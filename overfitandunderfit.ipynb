```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
def poly_reg(degree):
    poly_regr = Pipeline([("features",PolynomialFeatures(degree=degree)),
                         ("scaler",StandardScaler()),
                         ("regression",LinearRegression())])
    return poly_regr
```

```python
np.random.seed(600)
x = np.random.uniform(-3,3,size=100).reshape(-1,1)
y = 0.5*x**2 + x + 2 + np.random.normal(0,1,size=100).reshape(-1,1)
```

```python
from sklearn.metrics import mean_squared_error
lin_reg = LinearRegression()
lin_reg.fit(x,y)
plt.scatter(x,y)
plt.plot(x,lin_reg.predict(x),color = "red")
plt.show()
```

```python
mean_squared_error(y,lin_reg.predict(x))
```

```python
ploy2 = poly_reg(degree=2)
ploy2.fit(x,y)
y_predict = ploy2.predict(x)
plt.scatter(x,y)
plt.plot(np.sort(x,axis=0),y_predict[np.argsort(x,axis=0)].reshape(-1,1))
plt.show()
```

```python
mean_squared_error(y,ploy2.predict(x))
```

```python
ploy3 = poly_reg(degree=3)
ploy3.fit(x,y)
y_predict = ploy3.predict(x)
plt.scatter(x,y)
plt.plot(np.sort(x,axis=0),y_predict[np.argsort(x,axis=0)].reshape(-1,1),color = "red" )
plt.show()
```

```python
mean_squared_error(y,ploy3.predict(x))

```

```python
ploy100 = poly_reg(degree=100)
ploy100.fit(x,y)
y_predict = ploy100.predict(x)
plt.scatter(x,y)
plt.plot(np.sort(x,axis=0),y_predict[np.argsort(x,axis=0)].reshape(-1,1),color = "red" )
# plt.plot(np.sort(x,axis=0),ploy100.predict([np.argsort(x,axis=0)].reshape(-1,1)),color = "red")
plt.show()
```

```python
mean_squared_error(y,ploy100.predict(x))

```
