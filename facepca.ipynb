```python
random_index = np.random.permutation(len(faces.data))
randomdata = faces.data[random_index]
```

```python
data_example = randomdata[:36,:]
data_example.shape
```

```python
def plot_digits(data):
    fig,axes = plt.subplots(6,6,figsize = (10,10),
                            subplot_kw={'xticks':[],'yticks':[]},
                            gridspec_kw=dict(hspace = 0.1,wspace = 0.1))
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(62,47),cmap='bone')
```

```python
plot_digits(data_example)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(svd_solver="randomized")
pca.fit(randomdata)
```

```python
pca.components_.shape
```

```python
plot_digits(pca.components_[:36,:])

```
